```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
import os
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
```

```python
current_path=os.getcwd()
#print(current_path)
customer_data = pd.read_csv(current_path+'/shopping_data.csv')
```

```python
#show the data
customer_data.shape
customer_data.head()
```

```python
#We can remove CustomerID column, Genre, and Age column. We will retain the Annual Income (in thousands of dollars) and Spending Score (1-100) columns. The Spending Score column signifies how often a person spends money in a mall on a scale of 1 to 100 with 100 being the highest spender.
data = customer_data.iloc[:, 3:5].values
```

```python
plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
```
